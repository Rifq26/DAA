```python
# Hitung Inversi
# Inversi dalam suatu array

def countInversi(arr):
    result=0
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            if arr[i]>arr[j]:
                result+=1
    return result
```

```python
#arr = [21, 70, 36, 14, 25]
arr = [21, 70, 36]
```

```python
# mencetak berapa banyak dalam mengurutkan data
result = countInversi(arr)
print(result)
```

```python
# Inversi dalam suatu array
# Hitung Inversi dengan divide dan conquer

def countInversion(arr):
    icount = 0
    if len(arr)<=1: # len artinya length
        return icount
    
    mid = len(arr)//2
    # / menghasilkan koma (float)
    # // desimal (int)
    left = arr[:mid]
    right = arr[mid:]
    icount += countInversion(left)
    icount += countInversion(right)
    i=j=k=0
    
    #print(left)
    #print(right)
    while i<len(left) and j<len(right):
        if left[i] < right[j]:
            arr[k] = left[i]
            i+=1
        else:
            #print(Left[i],right[j])
            arr[k] = right[j]
            j+=1
            icount+=(mid-i)
        k+=1
        
    while i<len(left):
        arr[k] = left[i]
        i+=1
        k+=1
        
    while j<len(right):
        arr[k] = right[j]
        j+=1
        k+=1
        
    return icount
```

```python
#arr = [21, 70, 36, 14, 25]
arr = [1, 20, 6, 4, 5]
```

```python
result = countInversion(arr)
print(result)
```

```python
# Maximum Subarray Sum

# Tanpa devide dan conquer

def maxSubSum(arr):
    max_so_far = 0
    max_ending_here = 0
    for i in range(len(arr)):
        max_ending_here+=arr[i] # artiny max_ending_here = max_ending_here + arr[i]
        
        if max_ending_here > max_so_far:
            max_so_far = max_ending_here
        if max_ending_here < 0:
            max_ending_here = 0
        
    return max_so_far

arr = [-1, -5, 6, -2, -3, 1, 5, -6]
result = maxSubSum(arr)
print(result)
```

```python
# Maximum Subarray Sum

# Menggunakan devide dan conquer

def maxCrossingSum(arr,low,mid,high):
    result = 0; leftSum = float('-infinity')
    for i in range(mid,low-1,-1):
        result += arr[i]
        if result > leftSum:
            leftSum = result
    result = 0; rightSum=float('-infinity')
    for i in range(mid+1,high+1):
        result += arr[i]
        if result > rightSum:
            rightSum = result
    
    return leftSum + rightSum

def maxSum(arr,low,high):
    if low == high:
        return arr[low]
    mid = (low+high)//2
    
    return max(maxSum(arr,low,mid), maxSum(arr,mid+1,high), maxCrossingSum(arr,low,mid,high))
```

```python
arr = [-2, -5, 6, -2, -3, 1, 5, -6]
result = maxSum(arr,0,len(arr)-1)
print(result)
```

```python
# Longest Common Prefix

# Dengan divide dan conquer

def commonPrefix(str1, str2): # str yaitu String
    n1 = len(str1); n2 = len(str2)
    i,j = 0,0
    s = ""
    
    while i < n1 and j < n2:
        if str1[i] == str2[j]:
            s += str1[i]
            i += 1
            j += 1
        else:
            break
            
    return s

# devide conquer
def longestCommonPrefix(arr,low,high):
    if low == high:
        return arr[low]
    mid = (low+high)//2
    
    result1 = longestCommonPrefix(arr,low,mid)
    result2 = longestCommonPrefix(arr,mid+1,high)
    result = commonPrefix(result1,result2)
    return result

arr = ['geeksforgeeks', 'geeks', 'geek', 'geezer']
result = longestCommonPrefix(arr,0,len(arr)-1)
print(result)
```

```python
# Median dua array urut berbeda ukuran

# Floor in sorted array

def floorSorted(arr,low,high,x):
    if low > high:
        return -1
    
    if arr[low] > x:
        return -1
    
    if arr[high] <= x:
        return arr[high]
    
    mid = (low+high)//2
    
    if arr[mid] == x:
        return arr[mid]
    
    if mid > 0 and x >= arr[mid-1] and arr [mid] > x:
        return arr[mid-1]
    if mid < high and x < arr[mid+1] and x >= arr[mid]:
        return arr[mid]
    
    if x > arr[mid]:
        return floorSorted(arr,mid+1,high,x)
    else:
        return floorSorted(arr,low,mid-1,x)
    
arr = [1,2,8,10,12,14,19]
x = 5
```

```python
# Nilai terdekat

# Mencari nilai terdekat dengan metode devidi dan conquer
def closestNumber(arr,low,high,x):
    if low > high:
        return -1
    if arr[high] <= x:
        return arr[high]
    if arr[low] >= x:
        return arr[low]
    mid = (low+high)//2
    
    # abs yaitu absolute sehingga tidak menghasilkan tanda plus atau min, disebut nilai mutlak dari suatu bilangan riil
    if arr[mid] == x:
        return arr[mid]
    abs_mid = abs(arr[mid]-x)
    
    if mid > 0:
        abs_left = abs(arr[mid-1]-x)
        if abs_left < abs_mid:
            return closestNumber(arr,low,mid-1,x)
    if mid < high:
        abs_right = abs(arr[mid+1]-x)
        if abs_right < abs_mid:
            return closestNumber(arr,mid+1,high,x)
    #print('after')
    return arr[mid]

arr = [2, 5, 6, 7, 8, 8, 9]
#mencari nilai terdekat dari nilai yg ada di variabel x
x = 9
```

```python
print(closestNumber(arr,0,len(arr)-1,x))
```

```python
# Fixed Point

#Mencari nilai fixed point, yaitu suatu
#bilangan dalam array yang nilainya
#sama dengan urutannya dalam array. Array dimulai dari 0. Jika tidak ada
#yang sama, hasil akan bernilai -1.

# Mencari Fixed Point dengan metode divide dan conquer

def fixedPoint(arr,low,high):
    if low > high:
        return -1
    if arr[high] == high:
        return arr[high]
    if arr[low] == low:
        return arr[low]
    mid = (low+high)//2
    
    if arr[mid] == mid:
        return arr[mid]
    if mid > arr[mid]:
        return fixedPoint(arr,mid+1,high)
    else:
        return fixedPoint(arr,low,mid-1)
    
arr = [9, 1, 4, 5, 2]
```

```python
print(fixedPoint(arr,0,len(arr)-1))
```
