```python
import pulp
```

```python
#instantiate our problem class
model = pulp.LpProblem("ProfitMiximisingProblem", pulp.LpMaximize)
```

```python
A = pulp.LpVariable('A', lowBound=0, cat='Integer')
B = pulp.LpVariable('B', lowBound=0, cat='Integer')
```

```python
# mendefinisikan fungsi tujuan dan kendala sebagai berikut
# Objective function
model += 5000 * A + 2500 * B, "Profit"

# Contraints
model += 3 * A + 2 * B  <= 20
model += 4 * A + 3 * B  <= 30
model += 4 * A + 3 * B  <= 44
```

```python
# Solve our problem
model.solve()
pulp.LpStatus[model.status]
```

```python
# print our decision variable values
print (A.varValue)
print (B.varValue)
```

```python
# print our objective function value
print (pulp.value(model.objective))
```

# Latihan
Carilah nilai x dan y dari persamaan berikut


4x + 3y = 34

5x + y  = 37

```python
# Latihan

model = pulp.LpProblem("ProfitMiximisingProblem", pulp.LpMaximize)
X = pulp.LpVariable('X', lowBound=0, cat='Integer')
Y = pulp.LpVariable('Y', lowBound=0, cat='Integer')
model += 4 * X + 3 * Y  == 34
model += 5 * X + 1 * Y  == 37

model.solve()

print (X.varValue)
print (Y.varValue)
```

# Latihan
Budi membeli tiga pensil dan empat buku di toko Rana dengan harga Rp 11000,-. Jika Budi membeli lagi sebuah pensil dan tujuh buku ditoko yang sama dengan harga Rp 15000,-. Berapakah harga dua buah pensil dan enam buah buku jika Budi membeli kembali di toko Rana!

```python
# Latihan
model = pulp.LpProblem("ProfitMiximisingProblem", pulp.LpMaximize)
pensil = pulp.LpVariable('pensil', lowBound=0, cat='Integer')
buku = pulp.LpVariable('buku', lowBound=0, cat='Integer')

model += 3 * pensil + 4 * buku  == 11000
model += 1 * pensil + 7 * buku  == 15000

# harga 2 pensil dan 6 buku
model += 2 * pensil + 6 * buku, 'profit'
#model += 2 * pensil + 6 * buku

model.solve()

print (pensil.varValue)
print (buku.varValue)
print (pulp.value(model.objective))
```

# Latihan
Keliling sebuah persegi panjang sama dengan 44 cm. Jika lebarnya 6 cm lebih pendek dari panjangnya, carilah panjang dan lebar dari persegi panjang tersebut.

```python
# Latihan
model = pulp.LpProblem("ProfitMiximisingProblem", pulp.LpMaximize)
P = pulp.LpVariable('P', lowBound=0, cat='Integer')
L = pulp.LpVariable('L', lowBound=0, cat='Integer')

model += 2 * P + 2 * L, 'total'

model += 2 * P + 2 * L == 44
# Jika lebarnya 6 cm lebih pendek dari panjangnya
model += 2 * P + 2 * (P-6) == 44


model.solve()

print (P.varValue) #panjang
print (L.varValue) #lebar

print (pulp.value(model.objective)) #total keliling
```

# Menggunakan Scipy
Persamaan max z = 5x1 + 7x2 dimana

• 1x1 + 0x2 <= 16,

• 2x1 + 3x2 <= 19,

• 1x1 + 1x2 <= 8,

• x1, x2 >= 0

```python
# import required libraries
import numpy as np
from scipy.optimize import linprog
```

```python
# Set the inequality constraints matrix
# Note: the inequality constraints must be in the form of <=
A = np.array([[1,0], [2,3], [1,1], [-1,0], [0,-1]])

# Set the inequality constraints vextor
b = np.array([16, 19, 8, 0, 0])

# Set the coefficients of the linear objective function vector
# Note: when maximizing, change the signs of the c vector coefficients
c = np.array([-5, -7])
```

```python
# Solve linear proggraming (linprog) problem
res = linprog(c, A_ub=A, b_ub=b)
```

```python
# print results
print('Optimal value:', round(res.fun*-1, ndigits=2),
     '\nx.value:', res.x,
     '\nNumber of iterations performed:', res.nit,
     '\nStatus:', res.message)
```

# Latihan, library Scipy
Lisa dan Muri bekerja pada pabrik tas. Lisa dapat meyelesaikan 3 buah tas setiap jam dan Muri dapat menyelesaikan 4 tas setiap jam. Jumlah jam kerja Lisa dan Muri adalah 16 jam sehari dengan jumlah tas yang dibuat oleh keduanya adalah 55 tas. Jika jam kerja keduanya berbeda, tentukan jam kerja mereka masing- masing

```python
# Latihan 

A = np.array([[1,1], [3,4], [-1,0], [0,-1]])

b = np.array([16, 55, 0, 0])

c = np.array([-1, -1])

# Solve linear proggraming (linprog) problem
res = linprog(c, A_ub=A, b_ub=b)

# print results
print('Optimal value:', round(res.fun*-1, ndigits=2),
     '\nx.value:', res.x,
     '\nNumber of iterations performed:', res.nit,
     '\nStatus:', res.message)
```
