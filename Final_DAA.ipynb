```python
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt
import pandas as pd
```

```python
link_covid_data= 'https://github.com/owid/covid-19-data/raw/master/public/data/owid-covid-data.csv'
raw_data = pd.read_csv(link_covid_data) 
```

```python
indonesia = raw_data[raw_data['location'] == 'Indonesia'] 
turkey = raw_data[raw_data['location'] == 'Turkey']
poland = raw_data[raw_data['location'] == 'Poland']
argentina = raw_data[raw_data['location'] == 'Argentina']
```

```python
indonesia2 = indonesia[indonesia['date'] >= '2020-02-24']
turkey2 = turkey[turkey['date'] >= '2020-02-24']
poland2 = poland[poland['date'] >= '2020-02-24']
argentina2 = argentina[argentina['date'] >= '2020-02-24']
```

```python
def SelectionSort(list):
  for fill_slot in range(len(list) - 1,0,-1):
    max_index = 0
    for people_vaccinated in range(1, fill_slot + 1):
      if list[people_vaccinated] > list[max_index]:
        max_index = people_vaccinated
    list[fill_slot],list[max_index] = list[max_index],list[fill_slot]
  return list

list = [(indonesia2['people_vaccinated'])]
SelectionSort(list)
```
